```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def linear_function(x):
    y = 0.8 * x - 2
    return y

x_values = np.linspace(-15, 35, 20)
y_values = linear_function(x_values)
y_values = 0.8 * x_values - 2  # supported in np
# [1, 2, 3]
# 2 * x  - 5
# [-3, -1, 1, ]

plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # x-axis
plt.plot(x_values, y_values, color='green')
plt.title('y = 0.8x - 2')
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()

# targil
# x + y = 5
# 2x - y = 1
# line 1 - green
# line 2 - red
# add legend: with both equation
# junction point (2,3) -- # bonus use annotate to show the point
```
